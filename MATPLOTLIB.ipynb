```python
from matplotlib import pyplot as plt

stu_performance = ["Excellant", "Good", "Average", "poor"]
stu_value = [25,35,41,23]

plt.pie(stu_value,)

plt.show()
```

```python
#use labels, explode , shadow  , colour
stu_performance = ["Excellant", "Good", "Average", "poor"]
stu_value = [25,35,41,23]

plt.figure(figsize=(10,20)) #size of graph

plt.pie(stu_value,
        labels=stu_performance, #plot label
        explode=[0.2,0,0,0], #to join out the graph
        startangle=180,       #starting of angle from graph start
        colors=["r","b","k","orange"] #colors of the graph
        ,shadow=True)      #we want shadow or not

plt.legend(title="performance")

plt.show()
```
